```python
import dlib
import cv2
import matplotlib.pyplot as plt
```

```python
# NOT USING GPU
dlib.DLIB_USE_CUDA
```

```python
img = cv2.imread('../multi-faces.jpg',)
cnn_detector = dlib.cnn_face_detection_model_v1('../models/mmod_human_face_detector.dat')
detections = cnn_detector(img,3)
for face in detections:
    cv2.rectangle(img,(face.rect.left(),face.rect.top()),(face.rect.right(),face.rect.bottom()),(255,0,255),2)
    print(face.confidence)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
